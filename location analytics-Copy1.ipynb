## Churn

```python
df = merged_df
df['date'] = df['from_date']
df = df.sort_values(['user_id', 'date'])
df
```

```python
df_last = df.groupby(['user_id'])['date'].max().reset_index()
df['difference'] = df.groupby('user_id')['date'].diff().dt.days.fillna(0).astype(int)
df['difference'] = df.groupby('user_id')['date'].diff().dt.days.fillna(0).astype(int)
df_mean = df.groupby(['user_id'])['difference'].mean().reset_index()
df_mean.columns = ['user_id', 'mean_period']
```

```python
df_mean
```

```python
last_sale = df.date.max()
list_cust = df['user_id'].unique().tolist()
cl = len(df['user_id'].unique())
```

```python
df_app_final = pd.DataFrame(columns = ['date', 'difference','user_id']) 
```

```python
i=0
while(i<cl):
    
    globals()['df_cust%s' % i] = df[df['user_id']==list_cust[i]]
    
    stdev = globals()['df_cust%s' % i].std(axis = 0, skipna = True)
    kp = pd.DataFrame(stdev)
    kp = kp.transpose()
    
    df_app = pd.DataFrame(columns = ['date', 'difference','user_id']) 
    df_app = df_app.append(kp)
    df_app['user_id'] = list_cust[i]
    
    df_app_final = df_app_final.append(df_app)
    
    i = i+1
 
```

```python
df_app_final = df_app_final[['user_id','difference']]
df_app_final.rename({'user_id': 'user_id', 'difference': 'std'}, axis=1, inplace=True)

i = 0
list_cn = []
list_prd = []
```

```python
while (i<cl):
        
        
    num = (df_mean[df_mean['user_id']==list_cust[i]].mean_period).reset_index()
    x = num.iloc[0]['mean_period']
        
    if (last_sale > df[df['user_id']==list_cust[i]].date.max() + timedelta(days = x)):
       #(‘last sale date’ - ‘last purchase date’)/(‘mean purchase period’)
        if x == 0:
            skip_prd = 'no second purchase'
            list_cn.append(list_cust[i]) 
            list_prd.append(skip_prd)
        
        else:            
            skip_prd = ((last_sale - df[df['user_id']==list_cust[i]].date.max())/x).days
            list_cn.append(list_cust[i]) 
            list_prd.append(skip_prd)
        
    else:
        
        list_cn.append(list_cust[i]) 
        list_prd.append(0)
        
    print(i)
        
    i = i+1
    
    
    
data = {'user_id': list_cn, 
        'skipped_period': list_prd}  
new = pd.DataFrame.from_dict(data) 


df_final_merge = pd.merge(df_mean,df_last, how='left', left_on=["user_id"], right_on=["user_id"])
df_final_merge = pd.merge(df_final_merge,new, how='left', left_on=["user_id"], right_on=["user_id"])
df_final_merge = pd.merge(df_final_merge,df_app_final, how='left', left_on=["user_id"], right_on=["user_id"])
        
df_final_merge.to_csv('customer_trajectory.csv', index = False)
```

## Forecasting

```python
import pandas as pd
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
import statsmodels.api as sm
```

```python
df.date.max()
```

```python
df = merged_df
df['date'] = df['from_date']
df['count'] = 1
```

```python
df = df[['date', 'count']]
df['date'] = pd.to_datetime(df['date']).dt.date
df_days = df.groupby(['date'])['count'].sum().reset_index()
```

```python
df_days
```

```python
model=sm.tsa.statespace.SARIMAX(df_days['count'],order=(0, 1, 0),seasonal_order=(0,1,0,30))
results=model.fit()
df_days['forecast']=results.predict(start=(320),end=(346),dynamic=True)
```
